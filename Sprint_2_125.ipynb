**Description fo the features individually**

The features/varibales in the dataset are explained below:

- Tempo: The tempo of the song. The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, the tempo is the speed or pace of a given piece and derives directly from the average beat duration.

- Energy: Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. Higher the value more energetic the song.

- Danceability: Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. The value ranges from 0 to 1. Higher the value more suitable the song is for dancing.

- Loudness: Loudness values are averaged across the entire track. It is the quality of a song. It ranges from -60 to 0 DB. Higher the value, the louder the song.

- Valence: A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).

- Liveness: Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides a strong likelihood that the track is live.

- Acousticness: A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.

- Speechiness: Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audiobook, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks.

- Key: Key is the pitch, notes or scale of song that forms the basis of a song. 12 keys are ranging from 0 to 11.

**Plotting popular values of categorical features** 

Defining a function to create visualization of distrubtion of duplicate values of categorical varibale


````python
def plot_pop_values(df, col):
    """
    Plot the popular values in a column of a DataFrame.

    Parameters:
    - df: DataFrame object
    - col: Column name (string) to analyze and plot

    Returns:
    - None
    """

    # Get the duplicated values in the specified column
    duplicated_values = df.loc[df[col].duplicated(keep = False), col].value_counts()

    # Plot the popular values
    duplicated_values.plot.bar()

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=85)

    #title
    plt.title(f'Popular {col}s')

    # Display the plot
    plt.show()

# Example usage
# Assuming you have a DataFrame named 'df' and a column named 'genre'
plot_pop_values(df, 'artist')
````

## Creating Functions for graphs

**Function for plotting distribution of numerical feature/ frequency of values in a column'**

````python
def plot_distribution(df, col, threshold, bin_width = 0.01):
    """
    Plot the distribution of a specified column in a DataFrame.

    Parameters:
    - df: DataFrame object
    - col: Column name (string) to plot
    - threshold: Threshold value for annotation (float)

    Returns:
    - None
    """

    # Plot the histogram
    df[col].hist(bins=df.shape[0], width = bin_width)

    # Add a vertical line at the threshold value
    plt.axvline(threshold, c='yellow', label=f'Cutt off value: {threshold}')

    # Add labels and title
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.title(f'Distribution of {col}')

    # Display the plot
    plt.legend()
    plt.show()
````

**Function to plot tracks below a specified threshold for a given column**

````python
# Function to plot tracks below a specified threshold for a given column
def plot_tracks_below_threshold(data, column, threshold):
    """
    Plot tracks above a specified threshold for a given column.

    Parameters:
    - data: DataFrame object containing the data
    - column: Column name (string) to analyze and plot
    - threshold: Threshold value (float) to filter the data

    Returns:
    - None
    """

    # Filter the data based on the threshold
    filtered_data = data[data[column] < threshold]

    # Set the figure size
    plt.figure(figsize=(12, 10))

    # Plot the bar chart with horizontal bars
    plt.barh(filtered_data['track_name'], filtered_data[column])

    # Add labels and title
    plt.xlabel(column)
    plt.ylabel('Track Name')
    plt.title(f'Tracks with {column} < {threshold}')

    # adding legends 
    plt.axvline(threshold, c='yellow', label=f'Cutt off value: {threshold}')

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=90)

    # Add gridlines
    plt.grid(True, axis='x')

    # Display the plot
    plt.legend()
    plt.show()
````

Correlations between features


````python
# Drop the extra feature and Compute the correlation matrix
correlation_matrix = df.drop('duration_min',axis = 1).corr()


# Creating a heatmap of the correlation between features
fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(correlation_matrix, ax=ax, annot=True)
ax.set_title('Heatmap for the correlation between features')
plt.show()
````

Which features are strongly positively correlated?


> It can be seen from the heatmap that the strongly positively correlated features are loudness and energy



Using querying and data wrangling, assuming 0.6 threshold for strong positive correlation

````python
# Filter strongly positively correlated values
strongly_positively_correlated = correlation_matrix[(correlation_matrix != 1) & (correlation_matrix > 0.5)]

# Drop rows and columns with all NaN values
strongly_positively_correlated.dropna(how='all', axis=0, inplace=True)
strongly_positively_correlated.dropna(how='all', axis=1, inplace=True)

# Display the resulting DataFrame
display(strongly_positively_correlated)
````
